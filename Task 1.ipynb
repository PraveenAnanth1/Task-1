```python
import pandas as pd
```

```python
df =pd.read_csv('breast-cancer.csv')
```

```python
print("Dataset Information:\n")
df.info()
```

```python
print("\nFirst 5 Rows of the Dataset:\n")
df.head() 
```

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
print("\nTransformed Dataset:\n")
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

print("\nScaling Numerical Features...\n")

scaler = StandardScaler()

numerical_cols = df.columns[2:]  
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print("\nData After Scaling:\n")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

print("\nVisualizing Outliers...\n")

plt.figure(figsize=(14, 10))
sns.boxplot(data=df[numerical_cols])
plt.xticks(rotation=90)
plt.title("Boxplots for Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='diagnosis', data=df, palette='Set2')
plt.title('Distribution of Diagnosis (Malignant vs Benign)')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.show()
```

```python
corr = df[numerical_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
